# Write a Python program to display values of y when 'x' is an independent variable having values [-10, 10, step_size = 0.1] and:


* y = wx + b; where 'w' and 'b' are constants
* y = x^2
* y = 1/ (1 + e^(-x)); where ^ means 'to the power'.
* y = (e^x - e^(-x)) / (e^x - e^(-x))
* y = g(f(x)) where u = f(x) = wx + b and y = g(u) = 1/ (1 + e^(-u))
* y = g(f(x)) where f(x) = wx + b, u = f(x), and g(u) = (e^u - e^(-u)) / (e^u + e^(-u))
* y = g3(f3(v)) where f3(v) = w3y1 + w4y2 + b, w = f3(v), and g3(w) = 1/ (1 + e^(-w))
  * y1 = g1(f1(x)) where f1(x) = w1x + b1, u1 = f1(x), and g1(u1) = 1/ (1 + e^(-u1))
  * y2 = g2(f2(x)) where f2(x) = w2x + b2, u2 = f2(x), and g2(u2) = 1/ (1 + e^(-u2))

# Importing *Libraries*

```python
import matplotlib.pyplot as plt
import numpy as np
```

# Function for display the graph

```python
def plotFunction(x, y, title):
    plt.plot(x, y)
    plt.xlabel('X Co-ordinates')
    plt.ylabel('Y Co-ordinates')
    plt.title(title)
    plt.grid(1)
    plt.show()
```

# Defining Parameters

```python
w = 5
b = 2
```

# Taking X Co-Ordinates Values

```python
x = np.arange(-10, 10.1, 0.1)
```

# y = wx + b

```python
y = w * x + b
title = f"y = {w}x + {b}"
plotFunction(x, y, title)
```

# y = x^2

```python
y = x ** 2
title = f"y = x^2"
plotFunction(x, y, title)
```

# y = 1/ (1 + e^(-x))

```python
y = 1 / (1 + np.exp(-x))
title = f"y = 1/ (1 + e^(-x))"
plotFunction(x, y, title)
```

# y = (e^x - e^(-x)) / (e^x + e^(-x))

```python
y = (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
title = f"y = (e^x - e^(-x)) / (e^x + e^(-x))"
plotFunction(x, y, title)
```

# y = g(f(x)) where u = f(x) = wx + b and y = g(u) = 1/ (1 + e^(-u))

```python
u = w * x + b
y = 1 / (1 + np.exp(-u))
title = f"y = g(f(x)) , f(x) = wx + b , g(u) = 1/ (1 + e^(-u))"
plotFunction(x, y, title)
```

# y = g(f(x)) where u = f(x) = wx + b, and g(u) = (e^u - e^(-u)) / (e^u + e^(-u))



```python
u = w * x + b
y = (np.exp(u) - np.exp(-u)) / (np.exp(u) + np.exp(-u))
title = f"y = g(f(x)), f(x) = wx + b , g(u) = (e^u - e^(-u)) / (e^u + e^(-u))"
plotFunction(x, y, title)
```

# y = g3(f3(v)) where w = f3(v) = w3y1 + w4y2 + b, and g3(w) = 1/ (1 + e^(-w))
  
*   y1 = g1(f1(x)) where u1 = f1(x) = w1x + b1, and g1(u1) = 1/ (1 + e^(-u1))
*   y2 = g2(f2(x)) where u2 = f2(x) = w2x + b2, and g2(u2) = 1/ (1 + e^(-u2))



```python
w1 = 2
w2 = 4
w3 = 6
w4 = 8
b1 = 5
b2 = 7

u1 = w1 * x + b1
y1 = 1 / (1 + np.exp(-u1))

u2 = w2 * x + b2
y2 = 1 / (1 + np.exp(-u2))

u3 = w3 * y1 + w4 * y2 + b
y = 1 / (1 + np.exp(-u3))

title = f"y = g3(f3(x))"
plotFunction(x, y, title)
```
